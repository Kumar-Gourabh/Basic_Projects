# Question 1: What is the major reason people being kidnapped in each and every state?

```python
#Importing Relevant Dataset from resource related to kidnapping
kidnap=pd.read_csv("39_Specific_purpose_of_kidnapping_and_abduction.csv")
kidnap
```

```python
#Describing the numerical Features
kidnap.describe(include='all')
# For Kidnapping Data, Year Ranges from 2001 to 2010
```

```python
kidnap.shape
```

```python
# Let's see State-wise Sum of Kidnappings

## Need to remove all rows with Group Name as 'Kidnap - Total' because we are interested in individual reasons
kidnap_state=kidnap[kidnap['Group_Name']!='Kidnap - Total'].groupby(['Area_Name']).agg({"K_A_Female_Total":"sum","K_A_Male_Total":"sum","K_A_Grand_Total":"sum"}).sort_values(by='K_A_Grand_Total',ascending=False).reset_index()
kidnap_state
```

```python
print('Top 5 States In Kidnapping Cases are:\n')
kidnap_state['Area_Name'].head()
```

```python
#lets group by reason and count the total number of kidnappings
kidnap_reason=kidnap[kidnap['Group_Name']!='Kidnap - Total'].groupby(["Group_Name"]).agg({"K_A_Grand_Total":"sum"}).reset_index()
kidnap_reason['Percent']=kidnap_reason['K_A_Grand_Total']/(kidnap_reason['K_A_Grand_Total'].sum())*100
print(kidnap_reason)
print('\n\nTotal Cases of Kidnapping in the given dataset are:',int(kidnap_reason['K_A_Grand_Total'].sum()))
plt.figure(figsize=(15,5))
sns.barplot(data=kidnap_reason,x='Group_Name',y='K_A_Grand_Total')
plt.xlabel("Reasons for Kidnapping in India")
plt.ylabel('Count of Cases')
plt.xticks(rotation=60)
plt.show()
```

# The 3 main Reasons for Kidnappings in India are : 
# 1. Marriage
# 2. For Other Purposes
# 3. For Illicit Intercourse

```python
#lets group by state and reason and looks at the total number of kidnappings
kidnap_state_reason=kidnap[kidnap['Group_Name']!='Kidnap - Total'].groupby(["Area_Name","Group_Name"]).agg({"K_A_Grand_Total":"sum"})
kidnap_state_reason
```

```python
# Let's find for particular states: eg; Delhi, West Bengal, Jharkhand, U.P., Assam and Andhra Pradesh

list=['Uttar Pradesh','Andhra Pradesh','West Bengal','Delhi','Jharkhand','Assam']

for i in list:
    print(i,'\n',kidnap_state_reason.loc[i].sort_values(by='K_A_Grand_Total',ascending=False).reset_index(),'\n\n')
    print('Total Kidnap Cases in ',i,' are:',int(kidnap_state_reason.loc[i].reset_index()['K_A_Grand_Total'].sum()))
    plt.figure(figsize=(10,3))
    sns.barplot(data=kidnap_state_reason.loc[i].reset_index(),x='Group_Name',y='K_A_Grand_Total')
    plt.xlabel("Reasons for Kidnapping")
    plt.ylabel('Count of Cases')
    plt.xticks(rotation=60)
    plt.show()
    print('Top 3 Kidnap Cases in ',i,' are due to:\n',kidnap_state_reason.loc[i].sort_values(by='K_A_Grand_Total',ascending=False).reset_index().head(3),'\n\n')
    print("********************************************************************\n")
```

# Question 2: Offenders relation to the rape victim

```python
# We can use the 'Victims of rape' dataset to find whether the rape offenders are related to the victim amd what percent of rape cases are incest cases:

rape_victim=pd.read_csv('20_Victims_of_rape.csv')
rape_victim.head()
```

```python
# Let's have a look at State wise cases of Victims of Rape Victims
incest=rape_victim.groupby(['Area_Name','Subgroup']).agg({'Victims_of_Rape_Total':'sum'})
incest
```

```python
# Let's have a look at State wise cases of Victims of Rape Victims but this time, we'll see the percentage of Incest Rape Cases
incest.reset_index(inplace=True)
incest_total=incest[incest['Subgroup']=='Total Rape Victims']
incest_df=incest[incest['Subgroup']=='Victims of Incest Rape']
incest_total.drop('Subgroup',axis=1,inplace=True)
incest_df.drop('Subgroup',axis=1,inplace=True)
incest_total.reset_index(inplace=True)
incest_df.reset_index(inplace=True)
incest_total.drop('index',axis=1,inplace=True)
incest_df.drop('index',axis=1,inplace=True)
incest_df.rename(columns={'Victims_of_Rape_Total':'Victims_Incest'},inplace=True)
print(incest_df.head())
print(incest_total.head())
```

```python
# Let's Merge the above 2 dataframes
incest_merge=pd.merge(incest_total, incest_df, on='Area_Name')
```

```python
incest_merge.head()
```

```python
#Let's add another Column Percentage
incest_merge['Percentage_incest']=incest_merge['Victims_Incest']/incest_merge['Victims_of_Rape_Total']*100
incest_merge
```

```python
# Overall Percentage of Incest Cases
print('Overall Percentage of Incest Cases',incest_merge['Victims_Incest'].sum()/incest_merge['Victims_of_Rape_Total'].sum()*100)
```

```python
print('Economic data\n')
economic.head(10)
```

```python
# Let's find the aggregate sums of the 3 datasets based on states:
print('Education Grouped by State')
education_group=education.groupby(["Area_Name"]).agg({"Education_Above_Primary_but_below_Matric_or_Higher_Secondary":"sum","Education_Illiterate":"sum","Education_Matric_or_Higher_Secondary_&_above":"sum","Education_Total":"sum","Education_Upto_primary":"sum"})
education_group.head(10)
```

```python
print('Family Backgroud grouped by States')
family_group=family.groupby(["Area_Name"]).agg({"Family_back_ground_Homeless":"sum","Family_back_ground_Living_with_guardian":"sum","Family_back_ground_Living_with_parents":"sum","Family_back_ground_Total":"sum"})
family_group.head()
```

```python
# Let's merge the above 3:

juvenile=pd.concat([education_group,economic_group,family_group],axis=1)
juvenile.head()
```

# Question 4: Which state has more crime against children and women?

```python
#Lets import the data set relevant to the women
women=pd.read_csv("42_District_wise_crimes_committed_against_women_2001_2012.csv")
print(women.head())
#lets import the dataset of crime against children
children=pd.read_csv("03_District_wise_crimes_committed_against_children_2001_2012.csv")
print(children.head())
```

```python
#States with number of crimes against women
women_group=women.groupby(["STATE/UT"]).agg({"Rape":"sum","Kidnapping and Abduction":"sum","Dowry Deaths":"sum","Assault on women with intent to outrage her modesty":"sum","Insult to modesty of Women":"sum","Cruelty by Husband or his Relatives":"sum","Importation of Girls":"sum"})
women_group.head()
```

```python
#Let's add another feature - Total_Crimes
women_list=women_group.columns
women_group["total_crimes"]=0
for col in women_list:
    women_group["total_crimes"]=women_group["total_crimes"]+women_group[col]
women_group.head()
```

```python
women_group.reset_index(inplace=True)
```

```python
#We can take a look at the top 5 States in terms of Crimes against women
women_group_total=women_group[['STATE/UT','total_crimes']].sort_values(ascending=False,by='total_crimes')
women_group_total.head()
```

```python
plt.figure(figsize=(15,5))
sns.barplot(data=women_group_total.head(),x='STATE/UT',y='total_crimes')
```

```python
# Now For Children data. We can directly use the Total Column for ths dataset
children_group=children.groupby(["STATE/UT"]).agg({"Total":"sum"}).sort_values(by='Total',ascending=False).reset_index()
print("Top 5 States showing Crimes against children: \n",children_group.head())
```

```python
plt.figure(figsize=(15,3))
sns.barplot(data=children_group.head(),x='STATE/UT',y='Total')
```

# Question 5: Age group wise murder victim

```python
# Let's load the relevant dataset
murder=pd.read_csv("32_Murder_victim_age_sex.csv")
murder.head()
```

```python
# We can observe NaN values which might effect the operations. Let's replace them with 0
murder.fillna(0,inplace=True)
murder.head()
```

```python
list_age=['Victims_Upto_10_Yrs','Victims_Upto_10_15_Yrs','Victims_Upto_15_18_Yrs','Victims_Upto_18_30_Yrs',
          'Victims_Upto_30_50_Yrs','Victims_Above_50_Yrs']

count_age=[]
for i in list_age:
    count_age.append(murder[murder['Sub_Group_Name']!='3. Total'][i].sum())


# Created 2 Lists - 1 with Column heads and another with sums of cases
# Let's create a dataframe from it

df_age=pd.DataFrame([list_age,count_age])
df_age=df_age.transpose()

df_age.columns=['Age_Group','Count']
print(df_age)
```

```python
# Let's visualize it in a graph
plt.figure(figsize=(18,3))
sns.barplot(x='Age_Group',y='Count',data=df_age)
```

```python
# Thus, we conclude that the maximum victims belong to age group - 18-30 years; followed by 30-50

# Let's also look atAge group wise murder victims, grouped by sex

murder_sex_age=murder.groupby(["Sub_Group_Name"]).agg({"Victims_Above_50_Yrs":"sum","Victims_Upto_10_15_Yrs":"sum","Victims_Upto_10_Yrs":"sum","Victims_Upto_15_18_Yrs":"sum","Victims_Upto_18_30_Yrs":"sum","Victims_Upto_30_50_Yrs":"sum","Victims_Total":"sum"})
murder_sex_age
#From below table we come to know in every age group male victims are more than female vctims
```
